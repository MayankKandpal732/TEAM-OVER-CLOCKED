```python
import pandas as pd
from pathlib import Path

data = ['/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-BlackBoard-Screen/images','/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-BlackBoard-Sreen-Teacher'
        ,'/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-Discuss-2024-9-17','/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-Handrise-Read-write-2024-9-17'
        ,'/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-Stand-2024-9-17','/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-Talk-2024-9-17'
        ,'/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-Talk-Teacher-Behavior-2024-9-17','/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-Teacher-2024-9-17'
        ,'/kaggle/input/datasets/shreyasudaya/scb-05-dataset/SCB-Dataset/SCB5-Teacher-Behavior-2024-9-17'
       ]
```

```python
dat = {}
for pth in data:
    paths = [path.parts[-2:] for path in
         Path(pth).rglob('*.*')] 
    val,train=0,0
    for x,y in paths:
        if x=='val':
            val+=1
        else:
            train+=1
    dat[pth] = (val,train)
    print(f"{val}  {train}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os


labels = [os.path.basename(p.rstrip('/images')) for p in dat.keys()]
vals = [v[0] for v in dat.values()]
trains = [v[1] for v in dat.values()]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(14, 7))
ax.bar(x - width/2, trains, width, label='Train', color='#3498db')
ax.bar(x + width/2, vals, width, label='Val', color='#e74c3c')
ax.set_ylabel('Number of Images')
ax.set_title('Dataset Distribution: Train vs Validation')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
